```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
```

```python
sm
```

```python
sm['Percentage']=sm['Total']/5
sm
```

```python
sm['Average']=sm['Total']/5
sm
```

```python
def clac_grade(P):
    if P<=100 and P>=80:
        return 'A'
    elif P<80 and P>=60:
        return 'B'
    elif P<60 and P>=50:
        return 'C'
    elif P<50 and P>=40:
        return 'D'
    elif P<40 and P>=30:
        return 'E'
    elif P<0 and P>=30:
        return 'F'
    else:
        return 'X'
    
sm['Grade']=sm['Percentage'].apply(clac_grade)
sm
```

```python
temp=sm.head()
temp
```

```python


for sub in ['English','Hindi','Maths','Science','SST']:
            plt.figure(figsize=(10,6)) 
            plt.plot(sm['Name'],sm[sub],marker='o',label=sub)
            plt.title("Marks in different subjects")
            plt.xlabel('Name')
            plt.ylabel('Marks')
            plt.legend()
            plt.grid(True)
            plt.show()
```
